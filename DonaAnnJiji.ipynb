```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Plastic Waste Around the World.csv')
```

```python
#Pie chart
waste_distribution = df.groupby('Country')['Total_Plastic_Waste_MT'].sum().nlargest(10)
plt.figure(figsize=(8, 8))
waste_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=180)
plt.title('Plastic waste distribution of Top 10 countries')
plt.ylabel('')
plt.show()
```

```python
#Bar Graph
cwr_count = df['Coastal_Waste_Risk'].value_counts()
plt.figure(figsize=(10, 6))
cwr_count.plot(kind='bar', color='darkblue')
plt.title('Distribution of Coastal Waste Risk by Country', fontsize=14)
plt.xlabel('Coastal Waste Risk Level', fontsize=12)
plt.ylabel('Number of Countries', fontsize=12)
plt.show()
```

```python
#Scatter graph
plt.figure(figsize=(10, 6))
plt.scatter(df['Total_Plastic_Waste_MT'], df['Recycling_Rate'], color='black')
plt.xlabel('Total waste produced', fontsize=13)
plt.ylabel('Recycling rate', fontsize=13)
plt.title('Total waste VS Recycling rate', fontsize=16)
plt.grid(True)
plt.show()
```

```python
#Heatmap
corr_col = ['Total_Plastic_Waste_MT', 'Recycling_Rate', 'Per_Capita_Waste_KG']
corr_matrix = df[corr_col].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar=True)
plt.title('Correlation Heatmap of Plastic Waste and Recycling Metrics')
plt.show()
```
